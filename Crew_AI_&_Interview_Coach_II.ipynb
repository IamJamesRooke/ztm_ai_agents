```python
# import libraries
from crewai import Agent, Crew, Process, Task
from langchain_openai import ChatOpenAI
from IPython.display import display, Markdown
from crewai_tools import ScrapeWebsiteTool, SerperDevTool
```

```python
# define LLM
llm = ChatOpenAI(model='gpt-4o', temperature=0.5)
```

# User Inputs

```python
# call for inputs: interviewer, company, job position, job description
interviewer = input('Interviewer: ')
company = input('Company: ')
job_position = input('Job Position: ')
job_description = input('Job Description: ')
```

# Research Crew

## Research Agent

```python
# Create the Research AI Agent
research_agent = Agent(
    role = "Research Agent",
    goal = "contduct in-depth research",
    backstory = "As a research specialist, your mission is to uncover detailed and relevant information.",
    llm = llm,
    tools = [ScrapeWebsiteTool(), SerperDevTool()]
)
```

## Research Tasks

### Researching the Company

```python
# Define the task of researching the company.
research_company_task = Task(
    description = f"""
    Research about {company}, their position in the sector and recent industry trends.
    Focus on recent news and developments over the last 12 months.
    """,
    expected_output = f"""
    A comprehensive report in Markdown with bullet points about {company} and its industry.
    """,
    agent = research_agent
)
```

### Researching the Interviewer

```python
# Define the task of researching the one giving the interview.
research_person_task = Task(
    description = f"""
    Research about {interviewer} with a focus on their professional career.
    """,
    expected_output=f"""
    A comprehensive report in Markdown with bullet points about {interviewer}.
    """,
    agent = research_agent
)
```

## AI Coach and Interview Prep

```python
# Build the AI Interview coach agent
coach_agent = Agent(
    role = "Interview Coach",
    goal = f"Help the user prepare for the job interview for {job_position} by grading the relevance of the answer to the question.",
    backstory = """
    You are an energetic and empowering interview coach, much like Tony Robbins.
    You believe in unlocking potential and helping individuals perform at their best.
    Your coaching style is direct, motivating, and focuses on building confidence and articulating strengths effectively.
    You'll provide insightful feedback to help the user excel in their interview.
    """,
    llm = llm,
)
```

## Prepare list of questions

```python
## Prepare List of Questions
define_questions_task = Task(
    description = f"""
    Define a list of questions based on
    - the company {company},
    - the job description {job_description}, and
    - the interviewer {interviewer}.
    """,
    expected_output = f"""
    A list of 20 questions in Markdown format that cover
    - culture and team fit,
    - job position fit,
    - background and ways of working,
    - growth mindset
    """,
    agent = coach_agent,
    context = [research_company_task, research_person_task]
)
```

# Assemble the Crew

```python
crew = Crew(
    agents = [research_agent, coach_agent],
    tasks = [research_company_task, research_person_task, define_questions_task],
    verbose = True,
    process = Process.sequential
)

result = crew.kickoff({"topic": "Write a list of questions to prepare for the interview."})
```

# Exporting

```python
# Check with crew otcome.
display(Markdown(result.tasks_output[2].raw))
```

## Export each of the tasks output

```python
# Export the company report
company_report_file = f"{company}_report.txt"

with open(company_report_file, "w") as file:
    file.write(result.tasks_output[0].raw)

# Export interviewer report
interviewer_report_file = f"{interviewer}_report.txt"

with open(interviewer_report_file, "w") as file:
    file.write(result.tasks_output[1].raw)

# Export the interview questions
interview_questions_file = f"{job_position}_questions.txt"

with open(interview_questions_file, "w") as file:
  file.write(result.tasks_output[2].raw)
```

# Interview Prep

```python
# Import the questions from the directory
with open (interview_questions_file, "r") as file:
  questions = file.read()
```

### Interview Agent

```python
# AI Agent that asks a question
interviewer_agent = Agent(
    role = "Interviewer Agent",
    goal = f"Ask the user all questions from {questions}.",
    backstory = f"You are a job interview expert for companies like {company}.",
    llm = llm,
)
```

### Interview Task

```python
interview_prep_task = Task(
    description = f"Ask the user a question from {questions}",
    expected_output = f"Ask only ONE question at a time.",
    agent = interviewer_agent,
    human_input = True,
)
```

### Feedback Task

```python
feedback_task = Task(
    description = "Give the User feedback on their answer.",
    expected_output = """
    Markdown answer of
    - what was good,
    - what could be improved, and
    - how to take it to the next level.
    """,
    agent = coach_agent,
    context = [interview_prep_task],
    human_input = True,
)
```

# Interview Prep Crew

```python
# Assemble interview prep crew
crew = Crew(
    agents = [interviewer_agent, coach_agent],
    tasks = [interview_prep_task, feedback_task],
    verbose = True,
    process = Process.hierarchical,
    manager_llm = ChatOpenAI(model='gpt-4o', temperature=0),
    memory = True,
    planning = True,
)

result = crew.kickoff({"topic": f"Ask me all questions from {questions}."})
```
